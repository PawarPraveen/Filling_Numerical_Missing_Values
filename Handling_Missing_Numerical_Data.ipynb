```python
#Handling missing numerical data using mean, median and mode

import pandas as pd
import numpy as np

data = {
    'Bedrooms' : [2, 1, 4, np.nan, 5, np.nan ,8],
    'Bathrooms' : [np.nan, 3, 2, 5, np.nan, 4,1],
    'sq_foot' : [100, np.nan, 200, 500, np.nan, 300, np.nan],
    'Price': [200000, 300000, np.nan, 250000, 500000, 700000, 400000]
}

df = pd.DataFrame(data)
```

```python
df
```

```python
#mean -> its effect by the outlier so we use to fill the missing bedroom and bathroom 
#they are not containing outlier
#we call mean imputation

df['Bedrooms'] = df['Bedrooms'].fillna(df['Bedrooms'].mean())
df['Bathrooms'] = df['Bathrooms'].fillna(df['Bathrooms'].mean())
```

```python
# sqr_foot and price may contains outlier so its better to use median 

df['sq_foot'] = df['sq_foot'].fillna(df['sq_foot'].median())
df['Price'] = df['Price'].fillna(df['Price'].median())
```
